```python
#データの読み込み
olddata=pd.read_csv("FEI_PREF_241230101142.csv",encoding="shift-jis")
```

```python
olddata.head()
```

```python
#データの整形、作成
olddata=olddata.drop("/項目",axis=1)
olddata["調査年"]=olddata["調査年"].str.replace("年度","")
olddata["調査年"]=olddata["調査年"].astype("int64")
olddata["子ども(15歳未満)一人当たり教育費【千円/人】"]=olddata["D310311_教育費（都道府県財政）【千円】"]/olddata["A1301_15歳未満人口【人】"]
olddata["子ども(15歳未満)一人当たり児童福祉費【千円/人】"]=olddata["D3103033_児童福祉費（都道府県財政）【千円】"]/olddata["A1301_15歳未満人口【人】"]
olddata["老人(60歳以上)一人当たり老人福祉費【千円/人】"]=olddata["D3103032_老人福祉費（都道府県財政）【千円】"]/olddata["A1416_60歳以上人口【人】"]
```

```python
olddata.head()
```

```python
#予算割合分析用データの作成(調査年は後に加える)
data = pd.DataFrame()
data["地域"]=olddata["地域"]
data["総人口【人】"]=olddata["A1101_総人口【人】"]
data["地方債【％】"]=(olddata["D310115_地方債（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["社会福祉費【％】"]=(olddata["D3103031_社会福祉費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["老人福祉費【％】"]=(olddata["D3103032_老人福祉費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["児童福祉費【％】"]=(olddata["D3103033_児童福祉費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["農林水産業費【％】"]=(olddata["D310306_農林水産業費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["商工費【％】"]=(olddata["D310307_商工費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["土木費【％】"]=(olddata["D310308_土木費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["都市計画費【％】"]=(olddata["D3103085_都市計画費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["教育費【％】"]=(olddata["D310311_教育費（都道府県財政）【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
data["公営企業投資額【％】"]=(olddata["D5300_公営企業総投資額【千円】"]/olddata["D3103_歳出決算総額（都道府県財政）【千円】"])*100
```

```python
data.head(60)
```

```python
#固定効果法を適用したデータの作成
fedata=pd.DataFrame()
fedata["地域"]=data["地域"]
fedata["総人口【人】"]=data["総人口【人】"]-data.groupby("地域")["総人口【人】"].transform('mean')
fedata["地方債【％】"]=data["地方債【％】"]-data.groupby("地域")["地方債【％】"].transform('mean')
fedata["社会福祉費【％】"]=data["社会福祉費【％】"]-data.groupby("地域")["社会福祉費【％】"].transform('mean')
fedata["老人福祉費【％】"]=data["老人福祉費【％】"]-data.groupby("地域")["老人福祉費【％】"].transform('mean')
fedata["児童福祉費【％】"]=data["児童福祉費【％】"]-data.groupby("地域")["児童福祉費【％】"].transform('mean')
fedata["農林水産業費【％】"]=data["農林水産業費【％】"]-data.groupby("地域")["農林水産業費【％】"].transform('mean')
fedata["商工費【％】"]=data["商工費【％】"]-data.groupby("地域")["商工費【％】"].transform('mean')
fedata["土木費【％】"]=data["土木費【％】"]-data.groupby("地域")["土木費【％】"].transform('mean')
fedata["都市計画費【％】"]=data["都市計画費【％】"]-data.groupby("地域")["都市計画費【％】"].transform('mean')
fedata["教育費【％】"]=data["教育費【％】"]-data.groupby("地域")["教育費【％】"].transform('mean')
fedata["公営企業投資額【％】"]=data["公営企業投資額【％】"]-data.groupby("地域")["公営企業投資額【％】"].transform('mean')
fedata["調査年"]=olddata["調査年"]
```

```python
fedata.head(323)
```

```python
#ラグ付き変数の作成の準備
xdata=fedata.iloc[:,2:12]
yotherdata=fedata[["地域","調査年","総人口【人】"]]
```

```python
xdata=xdata.shift(-322)
```

```python
#ラグ付き変数の作成
lagfedata=pd.concat([yotherdata, xdata], axis=1)
lagfedata=lagfedata.dropna()
```

```python
#説明変数の相関行列
xdata.corr()
```

```python
#社会福祉費を説明変数から脱落させる
lagfedata=lagfedata.drop(["社会福祉費【％】"],axis=1)
lagfedata
```

```python
Y=lagfedata["総人口【人】"]
X=lagfedata.iloc[:,3:12]
```

```python
#OLSによる重回帰分析を行う
import statsmodels.api as sm
model = sm.OLS(Y, sm.add_constant(X))
results = model.fit()
print(results.summary())
print('p-values\n', results.pvalues)
```

```python
#VIF値の確認
from statsmodels.stats.outliers_influence import *
num_cols = model.exog.shape[1]
vifs = [variance_inflation_factor(model.exog, i) for i in range(0, num_cols)]
pdv = pd.DataFrame(vifs, index=model.exog_names, columns=["VIF"])
print(pdv)
```
